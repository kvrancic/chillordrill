```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('courses.csv')
```

```python
df['overall'] = df['overall'].str.extract('(\d+[.0-9]{0,2})').astype(float)
```

```python
df
```

```python
df = df.where(df['overall'] > 0).dropna()
```

```python
sns.histplot(df['overall'])
plt.title('Distribution of Overall Ratings')
plt.show()
```
